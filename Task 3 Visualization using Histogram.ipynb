```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming 'train.csv' is in the same directory
train_data = pd.read_csv('train.csv')
```

```python
# Display basic information about the dataset
print(train_data.info())

# Display the first few rows of the dataset
print(train_data.head())
```

```python
# Plot a histogram of ages
plt.figure(figsize=(10, 6))
plt.hist(train_data['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
# Plot a histogram of ages using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(train_data['Age'].dropna(), bins=20, kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```
